## Data Cleaning

### Creazione copia del dataset
Per preservare l'integrità del dataset originale, viene creata una copia su cui vengono eseguite le operazioni di pulizia dati.

```python
dfhotelClean = dfhotel.copy()
```

### Rimozione duplicati
In seguito all'analisi preliminare, si procede con l'eliminazione delle osservazioni duplicate che sono state identificate precedentemente nel dataset.

```python
dfhotelClean = dfhotelClean.drop_duplicates()
```

### Standardizzazione dei nomi delle variabili
Per rendere il dataset più coerente e facilmente utilizzabile nelle successive analisi, in particolare in ambiente SQL, si procede alla standardizzazione dei nomi delle colonne.
In particolare, vengono rimosse maiuscole, in quanto il formato snake case è già adottato.

```python
dfhotelClean.columns = dfhotelClean.columns.str.lower()
```

### Identificazione delle celle vuote nascoste (spazi)

Risulta necessario verificare la presenza di valori potenzialmente mancanti che non vengono riconosciuti come NaN.
In particolare, nei dataset importati da file CSV possono essere presenti celle contenenti solo spazi (" "), che non vengono interpretate automaticamente come valori nulli.

Per questo motivo, si procede a un controllo preliminare delle colonne testuali, al fine di individuare quante celle risultano vuote dopo la rimozione degli spazi iniziali e finali.

```python
for col in dfhotelClean.select_dtypes(include=["object", "string"]):
    print(col, (dfhotelClean[col].str.strip() == "").sum())
```

Dopo aver identificato la presenza di celle contenenti solo spazi, è necessario procedere alla pulizia effettiva del dataset, in modo da:
- uniformare i dati testuali
- eliminare spazi iniziali e finali
- convertire i valori vuoti in NaN, rendendoli riconoscibili nelle analisi successive.

```python
cols = dfhotelClean.select_dtypes(include=["object", "string"]).columns

for col in cols:
    dfhotelClean[col] = dfhotelClean[col].str.strip()
```

### Conversione delle stringhe vuote in valori mancanti

Dopo la rimozione degli spazi, le celle contenenti solo spazi diventano stringhe vuote (""), che devono essere convertite in valori mancanti (NaN).

```python
dfhotelClean.replace("", np.nan, inplace=True)
```

### Gestione dei valori mancanti
Nella fase di analisi esplorativa (EDA) è stata evidenziata la presenza di valori mancanti all’interno del dataset.  
A seguito delle operazioni di pulizia preliminare (rimozione degli spazi e conversione delle stringhe vuote in NaN), si procede a un’analisi più approfondita della distribuzione dei valori mancanti, al fine di definire la strategia di trattamento più appropriata per ciascuna variabile.

```python
dfhotelClean.isnull().sum()
```

Le variabili latitudine e longitudine presentano un numero limitato di valori mancanti (circa lo 0,6% del dataset).  
Considerata la ridotta incidenza, si è scelto di rimuovere le osservazioni incomplete al fine di garantire la qualità delle analisi successive.

```python
dfhotelClean = dfhotelClean.dropna(subset=["lat", "lng"])
```

Le variabili positive_review e negative_review contengono valori testuali standardizzati ("No Positive" e "No Negative") utilizzati per indicare l’assenza di una recensione rispettivamente positiva o negativa.

```python
dfhotelClean["positive_review"] = dfhotelClean["positive_review"].str.lower()
dfhotelClean["negative_review"] = dfhotelClean["negative_review"].str.lower()
```

```python
dfhotelClean[["positive_review", "negative_review"]].head(10)
```

### Creazione di Variabili Derivate (Feature Engineering)

In questa fase, vengono generate nuove variabili a partire da quelle esistenti per arricchire l'analisi. 
Verranno create tre variabili derivate:
- review_length, calcolata sommando il numero di parole delle recensioni positive e negative, così da misurare l'engagement totale dell'utente. Sarà utile per verificare se esiste una correlazione tra la loquacità del cliente e il punteggio assegnato.
- hotel_country, viene estratta la nazione dalla colonna hotel_address così da permettere di aggregare i risultati e confrontare le medie dei punteggi tra i diversi paesi europei rappresentati nel dataset.
- hotel_city, viene estratta dalla colonna hotel_address, utile per confrontare le performance degli hotel tra le diverse città europee presenti nel dataset.

```python
dfhotelClean["review_length"] = dfhotelClean["review_total_negative_word_counts"] + dfhotelClean["review_total_positive_word_counts"]
dfhotelClean[["review_total_negative_word_counts","review_total_positive_word_counts","review_length"]].head()
```

```python
dfhotelClean["hotel_country"] = dfhotelClean["hotel_address"].str.strip().str.split().str[-1]
dfhotelClean['hotel_country'] = dfhotelClean['hotel_country'].replace('Kingdom', 'United Kingdom')
dfhotelClean["hotel_country"].sample(10)
```

Al fine di comprendere la struttura della variabile hotel_address e individuare la posizione della città all’interno della stringa, si procede con un’analisi esplorativa degli indirizzi unici. In particolare, vengono estratte le ultime due componenti di ciascun indirizzo, ipotizzando che possano contenere informazioni rilevanti come città e paese.

```python
indirizzi_unici = dfhotelClean["hotel_address"].unique()
esplorazione = pd.Series(indirizzi_unici).str.split().str[-2:]
print(esplorazione.value_counts().head(20))
```

```python
citta_uK = dfhotelClean[dfhotelClean["hotel_address"].str.contains("United Kingdom")]
citta_uK["hotel_address"].str.split().str[-5:].head(10)
```

Dall'analisi emerge che il dataset è circoscritto a sei città europee: Londra, Parigi, Amsterdam, Vienna, Barcellona e Milano.

Poiché la struttura delle stringhe varia tra i diversi paesi, rendendo inaffidabile un'estrazione basata solo sulla posizione, si è deciso di utilizzare una funzione di mappatura testuale. Questo garantisce di associare correttamente ogni hotel alla propria città, indipendentemente dal numero di parole o dai codici postali presenti nell'indirizzo.

```python
lista_citta = ["London", "Paris", "Barcelona", "Milan", "Vienna", "Amsterdam"]

def trova_citta(indirizzo):
    for citta in lista_citta:
        if citta in indirizzo:
            return citta
    return "Altra città"

dfhotelClean["hotel_city"] = dfhotelClean["hotel_address"].apply(trova_citta)
dfhotelClean["hotel_city"].sample(10)
```
